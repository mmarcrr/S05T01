### Probability Sampling

Casos en los que todas las unidades de una población dada tienen la misma probabilidad de ser seleccionadas. Esta técnica incluye muestreo aleatorio simple, muestreo sistemático, muestreo por conglomerados y muestreo aleatorio estratificado.


#### muestreo aleatorio simple

El método de muestreo aleatorio simple selecciona muestras aleatorias de un proceso o población donde cada unidad tiene la misma probabilidad de ser seleccionada. Este es el método más directo de muestreo probabilístico.

```python
simple_random_sample = df.sample(n=100).sort_values(by='PATIENT_VISIT_IDENTIFIER')

# Save the sample mean in a separate variable
simple_random_mean = round(simple_random_sample['TEMPERATURE_DIFF'].mean(),3)

# View sampled data frame
simple_random_sample
```

## muestreo sistemático
El método de muestreo sistemático selecciona unidades en función de un intervalo de muestreo fijo (es decir, cada n-ésima unidad se selecciona de un proceso o población determinados). Este método de muestreo tiende a ser más eficaz que el método de muestreo aleatorio simple.

```python
def systematic_sampling(df, step):
    
    indexes = np.arange(0,len(df),step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample
    
# Obtain a systematic sample and save it in a new variable
systematic_sample = systematic_sampling(df, 3)

# Save the sample mean in a separate variable
systematic_mean = round(systematic_sample['TEMPERATURE_DIFF'].mean(),3)

# View sampled data frame
systematic_sample
```

#### muestreo por conglomerados
El método de muestreo por conglomerados divide la población en conglomerados de igual tamaño n y selecciona conglomerados cada T-ésima vez.


```python
def cluster_sampling(df, number_of_clusters):
    
    try:
        # Divide the units into cluster of equal size
        df['cluster_id'] = np.repeat([range(1,number_of_clusters+1)],len(df)/number_of_clusters)

        # Create an empty list
        indexes = []

        # Append the indexes from the clusters that meet the criteria
        # For this formula, clusters id must be an even number
        for i in range(0,len(df)):
            if df['cluster_id'].iloc[i]%2 == 0:
                indexes.append(i)
        cluster_sample = df.iloc[indexes]
        return(cluster_sample)
    
    except:
        print("The population cannot be divided into clusters of equal size!")
        
# Obtain a cluster sample and save it in a new variable
cluster_sample = cluster_sampling(df,5)

# Save the sample mean in a separate variable
cluster_mean = round(cluster_sample['TEMPERATURE_DIFF'].mean(),3)

# View sampled data frame
cluster_sample
```

#### Muestreo aleatorio estratificado

El método de muestreo aleatorio estratificado divide la población en subgrupos (es decir, estratos) y selecciona muestras aleatorias donde cada unidad tiene la misma probabilidad de ser seleccionada.

```python
# Import StratifiedShuffleSplit
from sklearn.model_selection import StratifiedShuffleSplit

# Set the split criteria
split = StratifiedShuffleSplit(n_splits=4, test_size=100)

# Perform data frame split
for x, y in split.split(df, df['AGE_ABOVE65']):
    stratified_random_sample = df.iloc[y].sort_values(by='PATIENT_VISIT_IDENTIFIER')

# View sampled data frame
stratified_random_sample

# Obtain the sample mean for each group
stratified_random_sample.groupby('AGE_ABOVE65').mean().drop(['PATIENT_VISIT_IDENTIFIER'],axis=1)
```
